```python
### Load in the libraries and the data
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('top 500 products - requests.csv')
data.head()
```

```python
data2 = pd.read_csv('Trial Suppliers.csv')
data2.head()
```

```python
data2.shape
```

```python
top500 = data['Substance Substance Name']
top500 = list(top500)
top500
```

```python
top500_substances_requests = data[['Substance Substance Name','Product Total Number of Requests']]
top500_substances_requests
```

```python
data2['flag'] = data2['substance_name'].isin(top500).astype(int)
data2
```

```python
top500_substances_requests = top500_substances_requests.rename(columns={'Substance Substance Name': 'substance_name'})
top500_substances_requests
```

```python
data2
```

```python
data2.loc[data2['flag'] == 1]
```

```python
df3 = pd.merge(data2, top500_substances_requests, on='substance_name',how="left")
df3
```

```python
df3.loc[df3['flag'] == 1]
```

```python
df3['Product Total Number of Requests'] = df3['Product Total Number of Requests'].fillna(0)
df3
```

```python
df3.flag.value_counts()
```

```python
df3.head()
```

```python
df_flag = df3[['company_id',"company_name",'flag']]
df_reqeust_score = df3[['company_id',"company_name",'Product Total Number of Requests']]
```

```python
df_flag.loc[df_flag['flag'] == 1]
```

```python
df_flag.to_csv('scoring_before_average.csv')
```

```python
request_score = df_reqeust_score.pivot_table(columns=['company_id','company_name'], values='Product Total Number of Requests', aggfunc='mean').mean().reset_index()

product_density = df_flag.pivot_table(columns=['company_id','company_name'], values='flag', aggfunc='mean').mean().reset_index()
```

```python
product_density.columns
```

```python

product_density = product_density.rename(columns={0: 'Product Density Score'})
request_score = request_score.rename(columns={0: 'Requests Density Score'})
```

```python
profile = product_density.merge(request_score,on=['company_id','company_name'])
profile
```

```python
profile.to_csv('Scores.csv')
```
