```python
### Load in the libraries and the data
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('top 500 products - requests.csv')
data.head()
```

```python
data2 = pd.read_csv('Trial Suppliers.csv')
data2.head()
```

```python
data2.shape
```

```python
top500 = data['Substance Substance Name']
top500 = list(top500)
top500
```

```python
data2['flag'] = data2['substance_name'].isin(top500).astype(int)
data2.head()
```

```python
data2.shape
```

```python
data2.loc[data2['flag'] == 1]
```

```python
data2.flag.value_counts()
```

```python
data2.head()
```

```python
df = data2[["company_name",'flag']]
df
```

```python
s = df.pivot_table(index='company_name', columns='company_name', values='flag', aggfunc='mean').mean()
s
```

```python
ds = pd.DataFrame(s)
```

```python
ds.head()
```

```python
ds.to_csv('results.csv')
```
