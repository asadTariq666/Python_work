```python
from sqlalchemy import func
from sqlalchemy import orm
import sqlalchemy as sa

company_mart = "vault.company_mart"
product_mart = "vault.product_mart"
negotiation_history_mart = "vault.negotiation_history_mart"
negotiation_mart = "vault.negotiation_mart"
product_substance_mart = "vault.product_substance_mart"

cas_list_agg = func.array_agg(product_substance_mart.c.cas_number).label("cas_list")

q = sa.select(
    company_mart.name.label("company"),
    company_mart.has_active_trial_supplier_membership.label("in_trial"),
    company_mart.has_signed_supplier_membership_contract.label("has_signed_up"),
    func.sum(func.coalesce(product_mart.count_of_product_name, 0)).label("company_top_500_product_count"),
    func.sum(func.coalesce(product_mart.total_number_of_requests, 0)).label("company_top_500_request_count"),
    func.sum(func.coalesce(negotiation_mart.number_of_requests, 0)).label("total_substance_request_count_based_on_company_substances"),
    func.sum(func.coalesce(negotiation_history_mart.total_number_of_offers, 0)).label("total_substance_offers_count_based_on_company_substances"),
    func.round((func.sum(func.coalesce(product_mart.total_number_of_requests, 0)) /
                func.sum(func.coalesce(product_mart.count_of_product_name, 0))).cast(sa.Numeric), 0).label("avg_requests_per_product"),
    func.case([(func.sum(func.coalesce(negotiation_mart.number_of_requests, 0)) == 0, -1)],
              else_=func.round((func.sum(func.coalesce(product_mart.total_number_of_requests, 0)) /
                                func.sum(func.coalesce(negotiation_mart.number_of_requests, 0))).cast(sa.Numeric), 2)).label("market_share_requests"),
    func.round((func.sum(func.coalesce(negotiation_history_mart.total_number_of_offers, 0)) /
                func.coalesce(func.sum(func.coalesce(negotiation_mart.number_of_requests, 1)), 1)).cast(sa.Numeric), 2).label("market_request_to_offer_cr"),
    cas_list_agg
).where(
    company_mart.has_signed_supplier_membership_contract &
    ~company_mart.is_lanxess &
    (company_mart.contract_sign_dates != '{null}')
).where(
    product_mart.company_id_hash_key == company_mart.company_id_hash_key
).where(
    product_substance_mart.product_id_hash_key == product_mart.product_id_hash_key
).where(
    negotiation_mart.product_substance_id_hash_key == product_substance_mart.product_substance_id_hash_key
).where(
    negotiation_history_mart.negotiation_id_hash_key == negotiation_mart.negotiation_id_hash_key
).group_by(
    company_mart.name,
    company_mart.has_active_trial_supplier_membership,
    company_mart.has_signed_supplier_membership_contract
).alias("z")
```
