# Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Load Data Set

```python
chilla = pd.read_csv('ChillaData.csv')
chilla.head()
```

## Changing Binary optionset data to Numeric data

```python
field = chilla['field_of_study']
vaccinated = chilla['Vaccinated'].map({'Yes': 1, 'No': 0})
gender =chilla['Gender'].map({'Male': 1, 'Female': 0})
glasses = chilla['Glasses'].map({'Yes': 1, 'No': 0})
marital = chilla['Marital Status'].map({'Yes': 1, 'No': 0})
area = chilla['Area'].map({'Urbun': 1, 'Rural': 0})
```

## Select axis 

```python
xs= field
ys= chilla['Coding Hours']
```

## LinePlot

```python
plt.figure(figsize=(14,6))
a =sns.lineplot(x=xs, y=ys, data =chilla, ci=None )
a.set(xlabel='field of study', ylabel='Coding Hours')
plt.show()
```

## Add title

```python
plt.figure(figsize=(14,6))
a =sns.lineplot(x=xs, y=ys, data =chilla, ci=None )
a.set(xlabel='field of study', ylabel='Coding Hours')
plt.title('field of study vs coding hours')
plt.show()
```

## Adding Limits

```python
xs= field
ys= chilla['Coding Hours']
plt.figure(figsize=(14,6))
a =sns.lineplot(x=xs, y=ys, data =chilla, ci=None )
a.set(xlabel='field of study', ylabel='Coding Hours')
plt.ylim(1)
plt.title('field of study vs coding hours')
plt.show()
```

# Set Styles

```python
xs= field
ys= chilla['Coding Hours']
plt.figure(figsize=(14,6))
sns.set_style("whitegrid")
a =sns.lineplot(x=xs, y=ys, data =chilla, ci=None )
a.set(xlabel='field of study', ylabel='Coding Hours')
plt.ylim(1)
plt.title('field of study vs coding hours')
plt.show()
```

## Size of Figure and dark theme

```python
xs= field
ys= chilla['Coding Hours']
plt.figure(figsize=(14,6))
sns.set_style(style='dark')
a =sns.lineplot(x=xs, y=ys, data =chilla, ci=None )
a.set(xlabel='field of study', ylabel='Coding Hours')
plt.title('field of study vs coding hours')
plt.show()
```
