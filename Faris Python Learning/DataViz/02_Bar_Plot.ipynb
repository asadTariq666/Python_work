# Import Libraries


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## BarPlot

```python
flowers = sns.load_dataset('iris')
sns.barplot(x='species', y='sepal_width', data =flowers)
plt.title('Flowers graph')
plt.show()
```

## BarPlot w.r.t Petal Length

```python
flowers = sns.load_dataset('iris')
sns.barplot(x='species', y='petal_length', data =flowers)
plt.title('Flowers graph')
plt.show()
```

## Adding Hue

```python
ship = sns.load_dataset('titanic')
sns.barplot(x='who', y='alone',hue='sex', data =ship)
plt.title('titanic graph')
plt.show()
```

## Adding Hue

```python
ship = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone',hue='who', data =ship)
plt.title('titanic graph')
plt.show()
```

## Changing order

```python
ship = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone',hue='who', data =ship, order=['female','male'])
plt.title('titanic graph')
plt.show()
```

## Changing Colours

```python
ship = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone',hue='who', data =ship, order=['female','male'], color='red')
plt.title('titanic graph')
plt.show()
```

## Adding Confidence Interval

```python
ship = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone',hue='who', data =ship, order=['female','male'], color='red', ci=None)
plt.title('titanic graph')
plt.show()
```

## Adding Palette

```python
ship = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone',hue='who', data =ship, order=['female','male'], color='red', ci=None, palette='pastel')
plt.title('titanic graph')
plt.show()
```

## Adding Estimator

```python
from numpy import mean
ship = sns.load_dataset('titanic')
sns.barplot(x='class', y='fare',hue='who', data =ship, estimator= mean)
plt.title('titanic graph')
plt.show()
```

## Estimator = Median

```python
import numpy
ship = sns.load_dataset('titanic')
sns.barplot(x='class', y='fare',hue='who', data =ship, estimator= median)
plt.title('titanic graph')
plt.show()
```

## Adding Saturation

```python
import numpy
ship = sns.load_dataset('titanic')
sns.barplot(x='class', y='fare',hue='who', data =ship, estimator= mean, saturation= 3)
plt.title('titanic graph')
plt.show()
```
