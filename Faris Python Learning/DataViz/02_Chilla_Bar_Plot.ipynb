# Import Libraries and reading dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
chilla = pd.read_csv('ChillaData.csv')
chilla.head()
```

## Changing Binary optionset data to Numeric data

```python
field = chilla['field_of_study']
vaccinated = chilla['Vaccinated'].map({'Yes': 1, 'No': 0})
gender =chilla['Gender'].map({'Male': 1, 'Female': 0})
glasses = chilla['Glasses'].map({'Yes': 1, 'No': 0})
marital = chilla['Marital Status'].map({'Yes': 1, 'No': 0})
area = chilla['Area'].map({'Urbun': 1, 'Rural': 0})
```

```python
# Checking data
field,vaccinated,gender,glasses,marital,area
```

## Changing order

```python

sns.barplot(x='Age', y=gender ,hue=marital,  data =chilla, order=['1-15','16-20','21-25','26-30','31-35','36-40','41-45','46-60',])
plt.show()
```

## Changing Colours

```python

sns.barplot(x='Age', y=gender ,hue=marital,  data =chilla, order=['1-15','16-20','21-25','26-30','31-35','36-40','41-45','46-60'],color='red')

plt.show()
```

## Adding Confidence Interval

```python

sns.barplot(x='Age', y=gender ,hue=marital,  data =chilla,ci=None, order=['1-15','16-20','21-25','26-30','31-35','36-40','41-45','46-60'],color='red')
plt.show()
```

## Adding Palette

```python
sns.barplot(x='Age', y=gender ,hue=marital, palette='pastel', data =chilla,ci=None, order=['1-15','16-20','21-25','26-30','31-35','36-40','41-45','46-60'],color='red')
plt.show()
```

## Adding Estimator

```python
from numpy import mean
from numpy import median

sns.barplot(x='Age', y=gender,data =chilla ,hue=marital, estimator=mean)
plt.show()
```

## Estimator = Median

```python
sns.barplot(x='Age', y=gender,data =chilla ,hue=marital, estimator=median)
plt.show()
```

## Adding Saturation

```python
sns.barplot(x='Age', y=gender,data =chilla ,hue=marital,saturation=.2 )
plt.show()
```

## Horizontal Barplot

```python

sns.barplot(x=gender, y='Age' ,hue=glasses,  data =chilla, order=['1-15','16-20','21-25','26-30','31-35','36-40','41-45','46-60',])
plt.show()
```

## Adding Facecolor, errColor, ErrWidth and lineWidth 

```python
plt.figure(figsize=(12,6))
sns.barplot(x=field, y=glasses,data =chilla, linewidth =1, facecolor=(1,.4,.3,.4), errcolor='.1',errwidth='.5')
plt.show()
```
