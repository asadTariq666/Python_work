```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
data = pd.read_csv('data_viz.csv')
data.head()
```

```python
p = sns.countplot(x='Gender', data=data, hue ='Location')
p.set_title('Plot 1: Gender on board')
plt.show()
```

```python
p = sns.countplot(x='Location', data=data, hue ='Gender')
p.set_title('Plot 2: origin of students')
plt.show()
```

```python
p = sns.countplot(x='Age', data=data, hue ='Location')
p.set_title('Plot 3: AGE GROUPS')
plt.show()
```
