### 1. Drop the 3 columns that contribute the least to the dataset. These would be the columns with the highest number of non-zero 'none' values. Break ties by going left to right in columns.
```python
# Count number of none in all columns of Dataframe
zero_count = []
for column_name in df.columns:
    column = df[column_name]
    # Get the count of Zeros in column 
    count = (column == 'none').sum()
    zero_count.append(count) 
    print('Count of "none" in column ', column_name, ' is : ', count)
zero_count    
```

### 2. Certain values in some of the columns contain unnecessary apostrophes ('). Remove the apostrophes.

```python
# Removing ' from whole dataframe
df = df.replace("'", '', regex=True)
df.head(20)
```

### 3. The checking_status column has values in 4 categories: 'no checking', '&lt;', '0&lt;=X&lt;200', and '&gt;=200'. Change these to 'No Checking', 'Low', 'Medium', and 'High' respectively.

```python
df.loc[df["checking_status"] == "no checking", "checking_status"] = 'No Checking'
df.loc[df["checking_status"] == "<0", "checking_status"] = 'Low'
df.loc[df["checking_status"] == "0<=X<200", "checking_status"] = 'Medium'
df.loc[df["checking_status"] == ">=200", "checking_status"] = 'High'
```

```python
df["checking_status"].unique()# Checking set of values in the column
```
