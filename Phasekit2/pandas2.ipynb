# Q2 Pandas and Numpy

```python
import pandas as pd
import numpy as np
from pandas import Series
from pandas import DataFrame
```

### Q 2.1
Remove rows with missing values for Price and Category

```python
data = {'Make': [1,2,3,4,5,np.nan,6,7,np.nan,8,9,10,np.nan],
        'Model': [11,12,np.nan,13,14,np.nan,15,16,np.nan,np.nan,17,np.nan,19],
        'Price': [np.nan,np.nan,6,40,50,6,np.nan,70,np.nan,8,90,10,np.nan],
        'Category': [np.nan,np.nan,'suv','sedan','sedan',np.nan,'suv','sedan','suv',np.nan,'sedan','suv','suv']}
df = pd.DataFrame(data)
print (df)
```

```python
df.dropna(subset=['Price', 'Category'],  thresh=1,inplace=True)
df
```

### Q 2.2
Replace NaN values of Price with average of cars in same category 

```python
df['Price'] = df['Price'].fillna(df.groupby('Category')['Price'].transform('mean'))
df = df.reset_index()
df
```

### Q 2.3
Replace missing category values by basing category on Price range 

```python
range_min = df.groupby(['Category'])['Price'].min()
range_max = df.groupby(['Category'])['Price'].max()
#range_min,range_max
list_of_keys = range_max.keys()
list_of_keys
j = 0
for i in range(len(df['Price'])):
    for j in range(len(range_max)):
        if (df['Price'][i] >= range_min[j] and df['Price'][i] <= range_max[j]):
            df.loc[df['Category'].isnull(),'Category'] = list_of_keys[j]
            j +=1
```
