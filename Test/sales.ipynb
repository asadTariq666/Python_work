# Sales Analysis for Adventure Works

## Importing Library and dataset

```python
import pandas as pd
df = pd.read_csv('Internet Sales.csv')
df.head()
```

### Checking names of data columns

```python
df.columns
```

### Extracting year from Order Date.

```python
df['OrderDate'] = pd.DatetimeIndex(df['OrderDate']).year
```

```python
df.head()
```

-----------------------------------------------------------------------------------------
### 1. What are the internet sales by Year for Each country Region?

```python
result = df.groupby(['CountryRegionCode','OrderDate'])['SalesAmount'].sum().reset_index(name ='Total Sales')
print(result)
```

----------------------------------------------------------------------
### 2. Generate summary statistics for Yearly Income, gender by Country Regions

```python
df.groupby(['CountryRegionCode','Gender'])['YearlyIncome'].describe()
```

----------------------------------------------------------------------
### 3. What is the average price of product category that generates the most sales.

```python
x = df.groupby(['ProductKey'])['SalesAmount'].sum().reset_index(name ='Sales')
x.sort_values(['Sales'], ascending = False)
```

Maximum sales are of Product 312. Average price of this product is calculate below.

```python
result2 = df.groupby(['ProductKey'])['SalesAmount'].mean().reset_index(name ='Average Sales')
result2= result2.sort_values(['Average Sales'], ascending = False)
result2.head(1)

```

### Average Sales of Product '312' is 3578.27

### 4. Does owning a home makes a difference in sales?


checking count of transactions for both House Owners and Non House Owners.

```python
df['HouseOwnerFlag'].value_counts()
```

```python
result3 = df.groupby(['HouseOwnerFlag'])['SalesAmount'].sum().reset_index(name ='Sales')
print(result3)
```

Owning a home does not make a lot of difference, the total count of home owners is 2.3 times the non home owners, and the sales almost show the same difference. Which indicates that the average sale for both Home owners and Non home owners is almost similar. But the number of Sales by home owners is quite high as compared to non home owners, which indicate that House Owners are more prone to make a sale, hence the revenue generated by Home owners is quite high as compared to Non House Owners.
