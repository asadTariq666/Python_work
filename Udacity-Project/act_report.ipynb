In Visualization, I discovered the behaviour of retweets and favourite counts over time and located that there's not solely a gradual increase within the counts over time but a rise within the variability, particularly for the favourite counts.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
twitter_data = pd.read_csv('twitter_archive_master.csv', encoding = 'utf-8')
twitter_data.date = pd.to_datetime(twitter_data.date)
twitter_data.set_index('date', inplace = True)
twitter_data[['favourite_count', 'retweet_count']].plot(style = '.', alpha = 0.2)
plt.title('Favourite and Retweet counts over Time')
plt.xlabel('Date')
plt.ylabel('Count')
```

A lot of the tweets have very high retweet or favourite counts as represented by the outliers in the scatter-plot above.

```python
labels = np.full(len(twitter_data.prediction.value_counts()), '', dtype = object)
labels[0] = 'Golden Retriever'
labels[1] = 'Labrador Retriever'
labels[2] = 'Pembroke'
labels[3] = 'Chihuahua'
labels[4] = 'Pug'
twitter_data.prediction.value_counts().plot(kind = 'pie', labels = labels)
plt.title('Division Dog Breeds')
```

Out of all the predictions the five most typical predictions were dog breeds and that they, together, accounted for around twenty fifth of the full variety of tweets within the dataset.
