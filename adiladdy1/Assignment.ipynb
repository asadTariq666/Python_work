`````python
# write your code here
# current, total
indicator_ids1 = ['NY.GDP.PCAP.CD',"BG.GSR.NFSV.GD.ZS", 'SP.POP.TOTL',"SP.POP.TOTL.FE.IN","SP.POP.TOTL.MA.IN",
                 'SP.RUR.TOTL.FE.ZS','SP.URB.TOTL.FE.ZS','SP.RUR.TOTL.MA.ZS','SP.URB.TOTL.MA.ZS',"NY.GDP.MKTP.PP.CD",'NY.GDP.MKTP.PP.KD']
country_codes = ['FRA','GBR','IRL']
my_dataframe1  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=5)
my_dataframe1
`````

`````python
trade_inservices=my_dataframe1.iloc[::7].values
gdp_ppp_current=my_dataframe1.iloc[1::7].values
gdp_ppp_constant=my_dataframe1.iloc[2::7].values
gdp_per_capita=my_dataframe1.iloc[3::7].values
pop_total=my_dataframe1.iloc[4::7].values
pop_female=my_dataframe1.iloc[5::7].values
pop_male=my_dataframe1.iloc[6::7].values



trade_inservices,gdp_ppp_current,gdp_ppp_constant,gdp_per_capita,pop_total
`````

`````python
trade_inservices_list=list(trade_inservices[0])+list(trade_inservices[1])+list(trade_inservices[2])
gdp_ppp_current_list=list(gdp_ppp_current[0])+list(gdp_ppp_current[1])+list(gdp_ppp_current[2])
gdp_ppp_constant_list=list(gdp_ppp_constant[0])+list(gdp_ppp_constant[1])+list(gdp_ppp_constant[2])
gdp_per_capita_list=list(gdp_per_capita[0])+list(gdp_per_capita[1])+list(gdp_per_capita[2])
pop_total_list=list(pop_total[0])+list(pop_total[1])+list(pop_total[2])
pop_female_list=list(pop_female[0])+list(pop_female[1])+list(pop_female[2])
pop_male_list=list(pop_male[0])+list(pop_male[1])+list(pop_male[2])

# trade_inservices_list,gdp_ppp_current_list,
# pop_male_list
`````

`````python
cols=my_dataframe1.columns.values
cols=list(cols)
cols
index_=[]
for i in my_dataframe1.index:
#     print(i[0])
    index_.append(i[0])
index_=list(set(index_))
index_
ind_list=[]
for i in index_:
    for j in cols:
        ind_list.append((i,j))
print(ind_list)

ind=pd.MultiIndex.from_tuples(ind_list, names=['Economy', 'Year'])
ind

data_g_p=pd.DataFrame({'trade_inservices':trade_inservices_list, 'gdp_ppp_current':gdp_ppp_current_list,
                       'gdp_ppp_constant':gdp_ppp_constant_list,'gdp_per_capita':gdp_per_capita_list,
                      'pop_total':pop_total_list,'pop_female':pop_female_list,'pop_male':pop_male_list})
data_g_p.values
new_record = pd.DataFrame(data_g_p.values ,
                          index = ind, 
                          columns = pd.MultiIndex.from_product([['Trade'], ['trade_inservices', 'gdp_ppp_current','gdp_ppp_constant',
                                                                                      'gdp_per_capita','pop_total','pop_female','pop_male']]))
new_record
`````
