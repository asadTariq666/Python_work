```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
read_data = pd.read_csv("student-mat_targetBinSuccess.csv")
df = pd.DataFrame(read_data)
df.head()
```

```python
figure, axes= plt.subplots(ncols=2,nrows=2,figsize= (20,15))

plt.sca(axes[0][0])
plt.scatter(df['G1'], df['G3'])
plt.title("Zusammenhang zwischen Noten von G1 & G3", fontsize= 15, fontweight='bold', horizontalalignment='center')

plt.xlabel("Note der G1", fontsize= 10) 
plt.xticks(range(0,21,1))

plt.ylabel("Note der G3", fontsize= 10)
plt.yticks(range(0,21,1))
##########################################################################################################
plt.sca(axes[0][1])
grouped_df= df.groupby("Fedu")["G3"].mean()

plt.bar(grouped_df.index,grouped_df, color= "grey")
plt.title("Einfluss vom Vaterbildung auf G3", fontsize= 15, fontweight='bold', horizontalalignment='center')

plt.xlabel("Bildungsgrad der Vater", fontsize= 10) 
plt.xticks(color='grey',fontsize='10',fontweight='bold',horizontalalignment='center')

plt.ylabel("Durchschnitt der Noten der SchÃ¼ler", fontsize= 10)
y_axis = range(0,21,1)
plt.yticks(y_axis, color='grey',fontsize='10',fontweight='bold',horizontalalignment='center')
###########################################################################################################
plt.sca(axes[1][0])
grouped_Fjob= df.groupby("Fjob")["G3"].mean()
fjob_gstd = df.groupby("Fjob")["G3"].std()

plt.bar(grouped_Fjob.index,grouped_Fjob, color= "grey",yerr=list(fjob_gstd), align='center',capsize=10)
plt.title("Einfluss vom Vaterbildung auf G3", fontsize= 15, fontweight='bold', horizontalalignment='center')

plt.xlabel("Bildungsgrad der Vater", fontsize= 10,horizontalalignment='center') 
plt.xticks(color='grey',fontsize='10',fontweight='bold',horizontalalignment='center')

plt.ylabel("Durchschnitt der Noten der SchÃ¼ler", fontsize= 10)
y_axis = range(0,21,1)
plt.yticks(y_axis, color='grey',fontsize='10',fontweight='bold',horizontalalignment='center')
##########################################################################################################
plt.sca(axes[1][1])
abscences_success = df.groupby(["absences"])["success"].sum()

plt.bar(abscences_success.index,abscences_success, color= "grey")
plt.title("Einfluss von Abwesenheit auf Erfolg", fontsize= 15, fontweight='bold', horizontalalignment='center')

plt.xlabel("Abwesenheit", fontsize= 10,horizontalalignment='center') 
plt.xticks(range(0,76,2),color='grey',fontsize='10',fontweight='bold',horizontalalignment='center')
plt.ylabel("Erfolg", fontsize= 10)
plt.yticks(color='grey',fontsize='10',fontweight='bold',horizontalalignment='center')
plt.savefig('test1.pdf')
```
