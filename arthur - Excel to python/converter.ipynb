```python
# Importing Libraries
import pandas as pd 
import numpy as np
```

## Reading the excel file in the dataframe.


```python
df = pd.read_excel("Data Example 3.xlsx")
df.head()
```

## Reading the excel file in other dataframe to add the updated columns.


```python
df_2 = pd.read_excel("Data Example 3.xlsx")
df_2.head()
```

## Dropping the Columns that are to be calculated

```python
df = df.iloc[:,2:9]
df.head()
```

## Creating separate lists for each column for data calculation

```python
Open = df['Open']
High = df['High']
Low = df['Low']
Last = df['Last']
ATR = df['ATR']
FractalSell = df['FractalSell']
FractalWk = df['FractalWk']
len(Open),len(High),len(Low),len(Last),len(ATR),len(FractalSell),len(FractalWk)
```

```python
# Creating new lists of same length as of dataframe and initializing them with -1
OpenTrade = [-1] *len(Open)
Stoploss = [-1]* len(Open)
TakeProfit = [-1] * len(Open)
CloseTrade = [-1] *len(Open)


## For loop will be executed for each row in the dataframe.
for i in range(len(Open)):
    if i==0: # first row initialization
        OpenTrade[i]=(Open[i])
        Stoploss[i]=(0)
        TakeProfit[i]=(OpenTrade[i] + ((OpenTrade[i] - 0) *2))
        CloseTrade[i]=(0)
        #print(OpenTrade,Stoploss,TakeProfit,CloseTrade)
        
    else: # when  i!=0 i.e. from row 2 to end of file
        #print(i,'i')

        # when Previous value of CloseTrade is not 0
        if CloseTrade[i-1] != 0:
            OpenTrade[i]=(Open[i])
            if (OpenTrade[i]<FractalWk[i-1]):
                Stoploss[i]=(FractalWk[i])
            else:
                Stoploss[i]=(FractalWk[i-1])
            if FractalSell[i-1] != 0:
                Stoploss[i]=(FractalSell[i-1])
            TakeProfit[i]=(OpenTrade[i] + ((OpenTrade[i] - Stoploss[i]) *2))

        elif CloseTrade[i-1] == 0: # when Previous value of CloseTrade is  0
            OpenTrade[i]=(OpenTrade[i-1])
            if FractalSell[i-1] != 0:
                Stoploss[i]=(FractalSell[i-1])
            else:
                Stoploss[i]=(Stoploss[i-1])
            TakeProfit[i]=(TakeProfit[i-1])

        # for values of CloseTrade
        if Low[i]<= Stoploss[i]:
            CloseTrade[i]=(Stoploss[i]-OpenTrade[i])
            
        elif High[i]> TakeProfit[i]:
            CloseTrade[i]=(TakeProfit[i]-OpenTrade[i])
        else:
            CloseTrade[i]=(0)

        #print(OpenTrade[i],Stoploss[i],TakeProfit[i],CloseTrade[i])
```

## Assigning the newly created lists to the dataframe Columns

```python
df_2['OpenTradeNew'] = OpenTrade
df_2['StoplossNew'] = Stoploss
df_2['TakeProfitNew'] = TakeProfit
df_2['Close TradeNew'] = CloseTrade
```

## Saving in to a csv file for comparison, if you want to owerwrite in to the existing columns just remove the 'New' from the column names

```python
df_2.to_csv('Calculated.csv')
```
