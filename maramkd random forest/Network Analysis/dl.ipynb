```python
import glob
import pandas as pd

# create an empty list to store the dataframes
dfs = []

# use glob to get a list of all .log files in the folder
file_list = glob.glob('pwnjutsu_dataset-network-json-player_11/*.log')

# loop over each file, read it into a dataframe, and append to the list
for file in file_list:
    df = pd.read_json(file, lines=True)
    dfs.append(df)

# concatenate all dataframes in the list into a single dataframe
final_df = pd.concat(dfs, ignore_index=True)
```

```python
df_dl = final_df[0:1000]
```
